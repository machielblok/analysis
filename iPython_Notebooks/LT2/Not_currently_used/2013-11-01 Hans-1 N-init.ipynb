```python
def analyse_dark_esr(folder):
    a = sequence.SequenceAnalysis(folder)
    a.get_sweep_pts()
    a.get_readout_results('ssro')
    a.get_electron_ROC()

    x = a.sweep_pts # convert to MHz
    y = a.p0.reshape(-1)[:]

    fig, ax = subplots(1,1)
    a.plot_result_vs_sweepparam(name='ssro', ax=ax)
    ax.set_ylim(0.65,1.05)

    fit_result = fit.fit1d(x, y, esr.fit_ESR_gauss, guess_offset,
        guess_amplitude, guess_width, guess_ctr,
        (3, guess_splitN), 
        do_print=True, ret=True, fixed=[4])
    plot_tools.plot_fit1d(fit_result, x, ax=ax, plot_data=False)

    fig.savefig(os.path.join(folder, 'darkesr_analysis.png'),
            format='png')

folder = r'd:\measuring\data\20131101\133955_PulsarDarkESR_hans-sil1-default'

guess_offset = 1
guess_ctr = 2.828
guess_splitN = 2.193e-3
guess_width = 0.2e-3
guess_amplitude = 0.2

analyse_dark_esr(folder)
```

## MBI pulse

```python
def fit_mbi_rabi(folder, guess_f, guess_A, guess_k, fixed=[]):
    a = mbi.MBIAnalysis(folder)
    a.get_sweep_pts()
    a.get_readout_results('adwindata')
    a.get_electron_ROC()
    ax = a.plot_results_vs_sweepparam(ret='ax')
    x = a.sweep_pts.reshape(-1)
    y = a.p0.reshape(-1)
    a.finish()
    
    f = fit.Parameter(guess_f, 'f')
    A = fit.Parameter(guess_A, 'A')
    k = fit.Parameter(guess_k, 'k')
    p0 = [f, A, k]
    fitfunc_str = '(1 - A) + A * exp(-kx) * cos(2pi fx)'
    
    def fitfunc(x): 
        return (1.-A()) + A() * np.exp(-k()*x) * \
            np.cos(2*np.pi*f()*(x))
    
    res = fit.fit1d(x,y,None, p0=p0, fitfunc=fitfunc, fitfunc_str=fitfunc_str, do_print=True, ret=True)
    plot_tools.plot_fit1d(res, linspace(x[0],x[-1],201), ax=ax,
                          plot_data=False)

folder = r'd:\measuring\data\20131101\170638_PulsarMBIElectronRabi_cal_slow_pi_hans-sil1-default'
fit_mbi_rabi(folder, 1./0.03, 0.8, 0)
```

## MBI calibration

```python
polarization = error.Formula()
P0,Pm1,Pp1,Pnone = sympy.symbols('P0 Pm1 Pp1 Pnone')
polarization.formula = (Pm1 - Pnone) / (Pm1 + P0 + Pp1 - 3*Pnone)

def mbi_calibration_analysis(folder):
    a = mbi.MBIAnalysis(folder)
    a.get_sweep_pts()
    a.get_readout_results('adwindata')
    x = a.sweep_pts.reshape(-1)
    x_names = a.g.attrs['sweep_pt_names']
    y = a.normalized_ssro.reshape(-1)
    u_y = a.u_normalized_ssro.reshape(-1)
    a.finish()

    fig, ax = subplots(1,1)
    ax.bar(range(len(x)), 1-y, yerr=u_y, width=0.7, align='center', ecolor='k',
           color='RoyalBlue')
    ax.set_xlabel('Readout transition')
    ax.set_ylabel('$P(m_s = 1)$')
    
    ax.set_xticks(range(len(x)))
    ax.set_xticklabels(x_names)
    
    for i,s in zip(range(4), [Pm1,P0,Pp1,Pnone]):
        polarization.values[s], polarization.uncertainties[s] = y[i], u_y[i]
        
    pol = float(polarization.value())
    u_pol = float(polarization.uncertainty())
    
    info = "Polarization of $m_I = -1$:\n P = {:.3f} $\pm$ {:.3f}".format(pol, u_pol)
    ymax = ax.get_ylim()[1]
    ax.text(2.-0.3, ymax*0.95, info, ha='left', va='top', size='large',
            bbox=dict(facecolor='white', alpha=0.5))
    
    fig.savefig(os.path.join(folder, 'MBI_readout.png'))
```

```python
folder = tb.data_from_time('20131101203634')
mbi_calibration_analysis(folder)
```
