# MBI Calibration

```python
polarization = error.Formula()
P0,Pm1,Pp1,Pnone = sympy.symbols('P0 Pm1 Pp1 Pnone')
polarization.formula = (Pm1 - Pnone) / (Pm1 + P0 + Pp1 - 3*Pnone)

def mbi_calibration_analysis(folder):
    a = mbi.MBIAnalysis(folder)
    a.get_sweep_pts()
    a.get_readout_results('adwindata')
    x = a.sweep_pts.reshape(-1)
    x_names = a.g.attrs['sweep_pt_names']
    y = a.normalized_ssro.reshape(-1)
    u_y = a.u_normalized_ssro.reshape(-1)
    a.finish()

    fig, ax = subplots(1,1)
    ax.bar(range(len(x)), 1-y, yerr=u_y, width=0.7, align='center', ecolor='k',
           color='RoyalBlue')
    ax.set_xlabel('Readout transition')
    ax.set_ylabel('$P(m_s = 1)$')
    
    ax.set_xticks(range(len(x)))
    ax.set_xticklabels(x_names)
    
    for i,s in zip(range(4), [Pm1,P0,Pp1,Pnone]):
        polarization.values[s], polarization.uncertainties[s] = y[i], u_y[i]
        
    pol = float(polarization.value())
    u_pol = float(polarization.uncertainty())
    
    info = "Polarization of $m_I = -1$:\n P = {:.3f} $\pm$ {:.3f}".format(pol, u_pol)
    ymax = ax.get_ylim()[1]
    ax.text(2.-0.3, ymax*0.95, info, ha='left', va='top', size='large',
            bbox=dict(facecolor='white', alpha=0.5))
    
    fig.savefig(os.path.join(folder, 'MBI_readout.png'))
    
def mbi_success_stats(folder):
    a = mbi.MBIAnalysis(folder)
    agrp = a.adwingrp('adwindata')
    mbi_cycles = agrp['MBI_cycles'].value
    mbi_attempts = agrp['MBI_attempts'].value
    mbi_time = agrp['MBI_time'].value / 1e3
    a.finish()
    
    mean_cycles = mbi_cycles.mean()
    std_cycles = mbi_cycles.std()
    median_cycles = median(mbi_cycles)
    
    mean_attempts = mbi_attempts.mean()
    std_attempts = mbi_attempts.std()
    median_attempts = median(mbi_attempts)
    
    mean_time = mbi_time.mean()
    std_time = mbi_time.std()
    median_time = median(mbi_time)

    fig, (ax1, ax2, ax3) = subplots(1,3,figsize=(16,4), sharey=False)
    ax1.hist(mbi_cycles, bins=arange(-0.5, max(mbi_cycles)+1.5),
             color=COLORS[0], edgecolor='None')
    ax1.set_xlabel('MBI cycles before success')
    ax1.set_ylabel('Occurences')
    ax1.set_title('Cycles')
    ax1.set_xlim(0.5, max(mbi_cycles)+1.5)
    
    ymax = ax1.get_ylim()[1]
    xmax = ax1.get_xlim()[1]
    ax1.text(xmax*0.95, ymax*0.95, 
        r'$\langle n \rangle = {:.1f}$, $\sigma = {:.1f}$'.\
            format(mean_cycles, std_cycles) + '\nmedian = ${:.1f}$'.format(median_cycles),
        size='large', bbox=dict(facecolor='white', alpha=0.5),
        va='top', ha='right')
    
    ax2.hist(mbi_attempts, bins=arange(0.5, max(mbi_attempts)+1.5),
             color=COLORS[2], edgecolor='None')
    ax2.set_xlabel('MBI attempts in successful cycle')
    ax2.set_title('Attempts')
    ax2.set_xlim(-0.5, max(mbi_attempts)+1.5)
    
    ymax = ax2.get_ylim()[1]
    xmax = ax2.get_xlim()[1]
    ax2.text(xmax*0.95, ymax*0.95, 
        r'$\langle n \rangle = {:.1f}$, $\sigma = {:.1f}$'.\
            format(mean_attempts, std_attempts),
        size='large', bbox=dict(facecolor='white', alpha=0.5),
        va='top', ha='right')
    
    ax3.hist(mbi_time, bins=arange(-0.5, int(10*median_time)+0.5),
             color=COLORS[1], edgecolor='None')
    ax3.set_xlabel('Time before success (ms)')
    ax3.set_title('Time')
    
    ymax = ax3.get_ylim()[1]
    xmax = ax3.get_xlim()[1]
    ax3.text(xmax*0.95, ymax*0.95, 
        r'$\langle t \rangle = {:.1f}$, $\sigma = {:.1f}$'.\
            format(mean_time, std_time) + '\nmedian = ${:.1f}$'.format(median_time),
        size='large', bbox=dict(facecolor='white', alpha=0.5),
        va='top', ha='right')
    
    fig.savefig(os.path.join(folder, 'MBI_success_stats.png'))
```

```python
folder = tb.latest_data('MBIFidelity')
mbi_calibration_analysis(folder)
mbi_success_stats(folder)
```
